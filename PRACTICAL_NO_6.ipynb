```python
#Retrieve and Print Seaborn Dataset Names
import seaborn as sns
sns.get_dataset_names()
```

```python
#Load and Display Head of Titanic Dataset
df=sns.load_dataset('titanic')
df.head()
```

```python
# Display Count of Null Values in Each Column of the DataFrame
df.isnull().sum()
```

```python
#To fill null values in the age column
mean_age=df['age'].mean()
df['Age Mean']=df['age'].fillna(mean_age)
df
```

```python
#To fill null values in the age column
median_age=df['age'].median()
df['Age Median']=df['age'].fillna(median_age)
df
```

```python
#To fill null values in the age column
mode_age=df['age'].mode()[0]
df['Age Mode']=df['age'].fillna(mode_age)
df
```

```python
#To drop a column
df_dropped=df.drop('Age Mean', axis=1)
df_dropped.head()
```

```python
df_dropped_multiple=df.drop(['Age Mode','parch'],axis=1)
df_dropped_multiple
```

```python
#To drop records
df_dropped_multiple=df.drop([2,3],axis=0)
df_dropped_multiple
```

```python
#Rename a column
df_renamed=df.rename(columns={'sex':'gender'})
df_renamed.head()
```
